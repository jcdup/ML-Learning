```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
mean = [0, 0]
cov = [[1, -0.5], [-0.5, 2]] 
```

```python
x, y =  np.random.multivariate_normal(mean, cov, 1000).T
```

```python
plt.plot(x, y, 'x')
#plt.axis('equal')
plt.show()
```
